```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import os
import regex as re

# Instantiate options
opts = Options()
# opts.add_argument(" — headless") # Uncomment if the headless version needed
# opts.binary_location = "<path to Chrome executable>"

# Set the location of the webdriver
chrome_driver = os.getcwd() + "/chromedriver"

# Instantiate a webdriver
driver = webdriver.Chrome(options=opts, executable_path=chrome_driver)

# Load the HTML page
driver.get("https://hello.iitk.ac.in/index.php/user/login")
```

```python
import csv

f=open(os.getcwd() + "/data.csv","w")
writer = csv.writer(f)
writer.writerow(["Week_No","Heading","LectureName","Lecture link"])
```

```python
# Parse processed webpage with BeautifulSoup
# Web Scrapping of the selected course page
#############Main driver code ############################

n=input("Enter how many lectures you want to fetch: ")
soup = BeautifulSoup(driver.page_source,"html.parser")
url_=driver.current_url
url_=url_.rsplit("/",2)[0]
print(url_)

weeks=soup.find_all('div',class_="weekItem")
weeks_content=soup.find_all('div',class_="weekDetailsBox")
week_name=[]
desc={}
lec_detail={}

def clean(s):
    s=re.sub("[^a-zA-Z0-9\\s]",'', s)
    s=s.strip()
    return s

for (week,details) in zip(weeks,weeks_content):
    week_no=week.find('div',class_="weekWrapper").text
    week_no=clean(week_no)
    week_name.append(week_no)
    
    list_=details.find_all('li')
    headingArr=[]
    for item in list_:  
        heading=item.find('span',class_="weekListItemTitle").text[2:]
        heading=clean(heading)
        headingArr.append(heading)
        
#         div_lec=item.find_all('span')
        anchors=item.find_all('a')
        lecDetail=[]
        for link in anchors:
            link_=link.get("href")
            if(link_.split('/',2)[0]!='#'):
                continue
            if(link_.split('/',2)[1]!='lecture'):
                continue
            link_=url_+'/'+link_
            print(link_)
            lecname=link.find('div',class_="lectureInfoBoxText").text
            print(lecname)
            lecname=clean(lecname)
            lecDetail.append([link_,lecname])
        lec_detail[heading]=lecDetail
    desc[week_no]=headingArr
```

```python
#insert data in csv
i=0
n=int(n)

for week in reversed(week_name):
    head=desc[week]
    
    for heading in reversed(head):
        lectures=lec_detail[heading]
        for lecture in reversed(lectures):
            row=[week,heading,lecture[1],lecture[0]]
            print(row)
            writer.writerow(row)
            i=i+1
            if(i == n):
                break
        if(i == n):
            break
    if(i == n):
        break

if(i<n):
    print("All latest lectures printed.No new lectures except this")
f.close()
```
